# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

# Visualization

```python
num_val = df.select_dtypes(include=np.number)
for col in num_val :
    sns.histplot(data=df,x = col , kde=True, bins=10)
    plt.show()
```

```python
plt.plot(df["Date"], df['Screen_Time (mins)'], marker='o', linestyle='-', color='b')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Visualization")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
sns.lineplot(x=df["Date"], y=df['Call_Duration (mins)'], marker='o')
plt.xticks(rotation=45)
plt.title("Seaborn Time Series Plot")
plt.show()
```

```python
sns.countplot(y=df['App_Usage (Top App)'],order= df['App_Usage (Top App)'].value_counts().index,palette='magma')
plt.title('The top app usage counts')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x=df['App_Usage (Top App)'],y=df['Battery_Level (%)'], data=df,palette='coolwarm')
plt.title('The app wise battery level')
plt.show()
```
